```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
#importing Dataset
CO2_data_path = "../Module_1_Assignment/co2.csv"
CO2_data_df = pd.read_csv(CO2_data_path)
# Displaying the data table for preview
CO2_data_df.head()
```

```python
CO2_data_df.info()
```

```python
CO2_data_df["CO2"].plot(kind= 'hist', bins = 6)
plt.xlabel('CO2 Levels')
plt.ylabel('Frequency')
plt.title('Histogram of CO2 Levels')
#plt.xticks(range(0, 10, 20))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Boxplot Plot
CO2_data_df["CO2"].plot(kind= 'box')
#plt.xlabel('CO2')
plt.ylabel('CO2 Levels')
plt.title('Box Plot of CO2 Levels')
```

```python
#Mean, standard deviation
mean_value = np.mean(CO2_data_df)
std_value = np.std(CO2_data_df)
```

```python
#Min, Q 1 , Q 2 , Q 3 , Max
min_value = np.min(CO2_data_df)
q1_value = np.percentile(CO2_data_df, 25)
median_value = np.median(CO2_data_df)
q3_value = np.percentile(CO2_data_df, 75)
max_value = np.max(CO2_data_df)
```

```python
from scipy.stats import binom

n = 20
p = 0.4
x = 7

probability_P_Y7= binom.pmf(x, n, p)
print("P(Y=7) =", probability_P_Y7)
```

```python
from scipy.stats import norm

mean = 20
std = 5
x = 15

probability_P_X_15 = norm.cdf(x, mean, std)
print("P(X ≤ 15) =", probability_P_X_15)
```

```python
Summary_of_Calclulation = pd.DataFrame ({"Mean:" : mean_value, 
"Standard Deviation:" : std_value, 
"Minimum:": min_value, 
"Q1:" : q1_value, 
"Q2:" : median_value, 
"Q3:" : q3_value, 
"Maximum:" : max_value,
"P(X ≤ 15)" : probability_P_X_15,
"P(Y=7)" : probability_P_Y7})
Summary_of_Calclulation
```
